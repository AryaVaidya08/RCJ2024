```python
import pandas as pd
import numpy as np
```

```python
letter = "U"
version = "4_27_24"
```

```python
#df = pd.read_csv("./images/version1/Labeled/H/Version_4_27_24/ImagesLabels.csv")
df = pd.read_csv(f"../../images/version1/Labeled/{letter}/Version_{version}/ImagesLabels.csv")
df
```

```python
imageStrs = df["image"]

fileNames = []
fileWidths = []
fileHeights = []
fileX_Min = []
fileY_Min = []
fileX_Max = []
fileY_Max = []

for image in imageStrs:
    imageName = image[image.rfind("/") + 1:]
    imageName = imageName[:imageName.index(".")]
    
    file = open(f"../../images/version1/Labeled/{letter}/Version_{version}/Annotations/{imageName}.xml", "r")
    #file = open(f"./images/version1/Labeled/H/Version_4_26_24/Annotations/{imageName}.xml", "r")
    lines = file.readlines()

    imageNum = imageName[imageName.index("-") + 1:]
    
    imageDim = [dim[dim.index(">") + 1 : dim.rfind("<")] for dim in lines[16:18]]
    width = int(imageDim[0])
    height = int(imageDim[1])
    
    imageBoundBox = [box[box.index(">") + 1 : box.rfind("<")] for box in lines[27:31]]
    x_min = int(imageBoundBox[0])
    y_min = int(imageBoundBox[1])
    x_max = int(imageBoundBox[2])
    y_max = int(imageBoundBox[3])


    fileNames.append(imageName)
    fileWidths.append(width)
    fileHeights.append(height)
    fileX_Min.append(x_min)
    fileY_Min.append(y_min)
    fileX_Max.append(x_max)
    fileY_Max.append(y_max)
```

```python
finalDF = pd.DataFrame({"image" : fileNames, 
              "width" : fileWidths,
              "height" : fileHeights,
              "x_min" : fileX_Min,
              "y_min" : fileY_Min,
              "x_max" : fileX_Max,
              "y_max" : fileY_Max,
              "label" : 2})

finalDF
finalDF.to_csv(f"../../images/version1/Labeled/{letter}/Version_{version}/BB_Images.csv")
```
