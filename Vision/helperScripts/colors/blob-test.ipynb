```python
import cv2
import numpy as np
import pandas as pd
```

```python
params = cv2.SimpleBlobDetector_Params()

params.filterByArea = True
params.minArea = 50
params.maxArea = 100000

params.filterByCircularity = True
params.minCircularity = 0.45

params.filterByInertia = True
params.minInertiaRatio = 0.4
```

#        How to know the colors
#### Red -->   R/G + R/B >= 3.7
#### Yellow -->  
#### Green --> G/B + G/R >= 2.7

```python
vid = cv2.VideoCapture(0)
detector = cv2.SimpleBlobDetector_create(params)

all_data = []

while True:
    _, frame = vid.read()
    keypoints = detector.detect(frame)



    if len(keypoints) != 0:
        for keypoint in keypoints:
            print("-" * 25)
            color = frame[int(keypoint.pt[1]), int(keypoint.pt[0])]
            b_g = color[0] / color[1]
            g_r = color[1] / color[2]
            b_r = color[0] / color[2]

            bg_r = (color[0] * color[1]) / color[2]
            gr_b = (color[2] * color[1]) / color[0]
            br_g = (color[0] * color[2]) / color[1]

            data = [color[0], color[1], color[2], b_g, g_r, b_r, bg_r, gr_b, br_g]
            all_data.append(data)
            #print(f"{keypoint.pt} & {keypoint.size} --> {color}")
            print(f"B --> {color[0]:.3f}\tG --> {color[1]:.3f}\tR --> {color[2]:.3f}")
            print(f"B:G --> {b_g:.3f}\tG:R --> {g_r:.3f}\tB:R --> {b_r:.3f}")
            print(f"BG:R --> {bg_r:.3f}\tGR:B --> {gr_b:.3f}\tBR:G --> {br_g:.3f}")
            
            
            
            
    frame_keypoints = cv2.drawKeypoints(frame, keypoints, np.array([]), (255, 0, 255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)


    frame[30:35, 30:35] = (0,0,0)
    cv2.imshow("Thing", frame)
    cv2.imshow("Normal", frame_keypoints)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

vid.release()
cv2.destroyAllWindows()
```

```python
b = []
g = []
r = []
b_g = []
g_r = []
b_r = []
bg_r = []
gr_b = []
br_g = []

for data in all_data:
    b.append(data[0])
    g.append(data[1])
    r.append(data[2])
    b_g.append(data[3])
    g_r.append(data[4])
    b_r.append(data[5])
    bg_r.append(data[6])
    gr_b.append(data[7])
    br_g.append(data[8])
```

```python
thing = pd.DataFrame({"Blue" : pd.Series(b),
              "Green" : pd.Series(g),
              "Red" : pd.Series(r),
              "Blue : Green" : pd.Series(b_g),
              "Green : Red" : pd.Series(g_r),
              "Blue : Red" : pd.Series(b_r),
              "Blue * Green : Red" : pd.Series(bg_r),
              "Green * Red : Blue" : pd.Series(gr_b),
              "Blue * Red : Green" : pd.Series(br_g)}).to_csv("./full-yellow-data.csv")
```
