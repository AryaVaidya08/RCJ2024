```python
import cv2
import numpy as np
import pandas as pd
```

```python
params = cv2.SimpleBlobDetector_Params()

params.filterByArea = True
params.minArea = 50
params.maxArea = 100000

params.filterByCircularity = True
params.minCircularity = 0.45

params.filterByInertia = True
params.minInertiaRatio = 0.4
```

```python
vid = cv2.VideoCapture(0)
detector = cv2.SimpleBlobDetector_create(params)

all_data = []

while True:
    _, frame = vid.read()
    keypoints = detector.detect(frame)



    if len(keypoints) != 0:
        for keypoint in keypoints:
            print("-" * 25)
            color = frame[int(keypoint.pt[1]), int(keypoint.pt[0])]
            blue = color[0]
            green = color[1]
            red = color[2]

            if (red / green) + (red / blue) >= 3.5:
                print("RED DETECTED!")
            elif (green / blue) + (green / red) >= 2.7:
                print("GREEN DETECTED!")
            elif (green * red) / blue >= 400:
                print("YELLOW DETECTED!")
            
            
    frame_keypoints = cv2.drawKeypoints(frame, keypoints, np.array([]), (255, 0, 255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

    cv2.imshow("Normal", cv2.resize(frame_keypoints, (384, 218)))

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

vid.release()
cv2.destroyAllWindows()
```
